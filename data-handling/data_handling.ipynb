# Preparing Data

- In this notebook, we will handle the products and policy data.
- The products data `amazon.csv` is downloaded from this [Kaggle dataset](https://www.kaggle.com/datasets/karkavelrajaj/amazon-sales-dataset).
- The policy file `policy.txt` is generated with ChatGPT.

```python
import pandas as pd
df = pd.read_csv('amazon.csv')
df.info()
```

```python
# removing duplicates
df = df[~df.product_id.duplicated()]
df.head()
```

```python
# keeping only needed columns
cols = ['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 
        'discount_percentage', 'rating', 'rating_count', 'about_product']

df = df[cols]
df.head()
```

```python
# removing rs symbol in order to convert to USD 
remove_rs_symbol = lambda x: float(x[1:].replace(',', ''))
df.discounted_price = df.discounted_price.apply(remove_rs_symbol)
df.actual_price = df.actual_price.apply(remove_rs_symbol)
df.head()
```

```python
# converting to USD
rs_to_usd = 0.012
df.discounted_price *= rs_to_usd
df.actual_price *= rs_to_usd
df.head()
```

```python
# resetting index
df = df.reset_index()
df.index
```

```python
from tqdm import tqdm

# converting csv file to txt file
def convert_df_to_txt(df, filename):
    with open(filename, "w") as file:
        for index, row in tqdm(df.iterrows()):
            line = (
                f'Product name is: {row.product_name} - From categories: {row.category}. '
                f'Description is: {row.about_product} - Has a rating of {row.rating} from {row.rating_count} users. '
                f'Price after {row.discount_percentage} discount is ${row.discounted_price}.\n'
            )
            file.write(line)


convert_df_to_txt(df, 'products.txt')
```

```python
# reading policy.txt file
with open('policy.txt', 'r') as file:
    policy = ''
    for line in tqdm(file):
        policy += line

print(f'Number of lines in the policy file: {len(policy)}')
```

```python
# visualizing policy.txt lines
for i, x in enumerate(policy.split('\n\n')):
    print(f'((({i})))')
    print(x)
    print('###')
```

```python
# As we have seen in the previous cell
# Only these indices are worth keeping
chosen_chunks_indices = [2, 3, 4, 5, 6, 7, 9, 10, 11, 13, 14, 15, 16, 18, 19, 20, 22]
all_policy_chunks = policy.split('\n\n')
policy_chunks = [all_policy_chunks[i] for i in chosen_chunks_indices]
policy_chunks
```

```python
# splitting preoducts.txt into chunks
# with 512 size limit, every single product will fit a single line
def split_into_chunks(filename, chunk_size=512):
    chunks = []
    with open(filename, "r") as file:
        current_chunk = ""
        for line in tqdm(file):
            if len(current_chunk) + len(line) > chunk_size:
                chunks.append(current_chunk.strip())
                current_chunk = line
            else:
                current_chunk += line
        if current_chunk:
            chunks.append(current_chunk.strip())
    return chunks


products_chunks = split_into_chunks('products.txt')
```

```python
print(f'Number of products chunks: {len(products_chunks)}')
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# preparing embedding model
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain_core.documents import Document

# convert chunks to documents
product_docs = [Document(page_content=chunk) for chunk in products_chunks]
policy_docs = [Document(page_content=chunk) for chunk in policy_chunks]
```

```python
from langchain_community.vectorstores import Chroma

# embedding
product_vectorstore = Chroma.from_documents(
    product_docs,
    embedding_model,
    persist_directory="../chatbot/vectorstores/chroma_products",
    collection_name='products'
)

policy_vectorstore = Chroma.from_documents(
    policy_docs,
    embedding_model,
    persist_directory="../chatbot/vectorstores/chroma_policy",
    collection_name='policy'
)

# save to disk
product_vectorstore.persist()
policy_vectorstore.persist()
```
